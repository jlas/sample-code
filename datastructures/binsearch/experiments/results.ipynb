```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot(df_):
    fig, ax = plt.subplots()
    cache = any(['cache' in c for c in df_.columns])
    cols = {c:'vEB' if 'veb' in c else 'Vanilla' for c in df_.columns}
    df_ = df_.rename(columns=cols)
    df_.plot(ax=ax, marker=".")
    ax.set_ylabel('L1 Cache Misses' if cache else 'Time Elapsed (s)')
    ax.set_xlabel('Number of Elements (Tree Size)')
    return fig
```

# Results 1

* 1M bench_reps

```python
df = pd.read_csv('results1.csv').set_index('nelements')
```

```python
fig = plot(df[['bs_L1_cache_misses', 'veb_L1_cache_misses']])
fig.suptitle('1M Searches')
```

```python
fig = plot(df[['bs_seconds_elapsed', 'veb_seconds_elapsed']])
fig.suptitle('1M Searches')
```
